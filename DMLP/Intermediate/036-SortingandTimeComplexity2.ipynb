```python
def bubble_sort_optimized(a):
    for i in range(len(a)):
        is_sorted = True
        
        for j in range(len(a) - i - 1):
            if a[j] > a[j+1]:
                is_sorted = False
                a[j], a[j+1] = a[j+1], a[j]
        
        if is_sorted:
            break
                
    return a
```

```python
def bubble_sort_visual_unoptimized(a):
    print(a)
    count = 0
    
    for i in range(len(a)):
        for j in range(len(a) - i - 1):
            count += 1
            
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                
            print(f"\t{a}")
        
        print("-"*25)
        print(a)
        
    print(f"I ran {count} times!")
                
    return a
```

```python
def selection_sort(a):
    for i in range(len(a) - 1):
        current_min = i
        
        for j in range(i + 1, len(a)):
            if a[current_min] > a[j]:
                current_min = j
        
        if current_min != i:
            a[current_min], a[i] = a[i], a[current_min]
            
    return a
```

```python
selection_sort([2,8,5,3,9,4,1])
```

```python
def insertion_sort(a):
    for i in range(1, len(a)):
        index_to_insert = i
        j = i - 1
        
        while j >= 0:
            if a[j] < a[index_to_insert]:
                break
                
            a[j], a[index_to_insert] = a[index_to_insert], a[j]
            index_to_insert = j
            j -= 1
    
    return a
```

```python
'''
ALGORITHM

    In the selection sort, first of all, we set the initial element as a minimum.
    Now we will compare the minimum with the second element. If the second element turns out to be smaller than the minimum, we will swap them, followed by assign to a minimum to the third element.
    Else if the second element is greater than the minimum, which is our first element, then we will do nothing and move on to the third element and then compare it with the minimum.
    We will repeat this process until we reach the last element.
    After the completion of each iteration, we will notice that our minimum has reached the start of the unsorted list.

CODE
'''
def selectionSort(lis):
    for i in range(len(lis)):
        min_index = i
        min_str = lis[i]
        for j in range(i+1,len(lis)):
            if min_str>lis[j]:
                min_str = lis[j]
                min_index = j
        temp = lis[i]
        lis[i] = lis[min_index]
        lis[min_index] = temp
    return lis


```

##Q8. k iterations of insertion sort

Problem Description
Given a list of numbers, perform k iterations of insertion sort on it and return the element present at the center.

Note:
For a list of length n, return the n//2th indexed element after applying k iterations of insertion sort.

Input Format
The first line contains space separated values of Integers, arr
The value of k



Output Format
The element at center after k iterations of insertion sort.



Example Input
12 11 13 5 6
2



Example Output
13



Example Explanation
The list after 2 iterations of insertion sort becomes [11, 12, 13, 5, 6]. So the central element is 13.


```python
'''
ALGORITHM

    Iterate from arr[1] to arr[k] over the array, because we need k iterations.
    Compare the current element (key) to its predecessor.
    If the key element is smaller than its predecessor, compare it to the elements before.
    Move the greater elements one position up to make space for the swapped element.
    After k iterations return the middle element in the array, which is given by arr[len(arr)//2].
'''

CODE

def insertionSort(arr,k):
    for i in range(1, k+1):

        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key

    res= arr[len(arr)//2]
    return res
```
