```python
class Solution:
  def solve(self,A,B):
    result = 0
    A.sort()
    for i in range(B):
      if A[i] <= B:
        result += 1
        print(A[i])
    return result

A = [1,3,4,4,6]
B = 4
pt1 = Solution()
res = pt1.solve(A,B)
print("Result: ",res)
```
